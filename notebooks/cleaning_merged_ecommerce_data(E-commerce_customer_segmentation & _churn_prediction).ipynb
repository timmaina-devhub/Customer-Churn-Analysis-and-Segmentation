**Based on findings above;Customers who have negative amounts should have refunded status else completed**

```python
data_sub['status'].value_counts()
```

```python
amounts=data_sub['amount']
status=data_sub['status']

status_redone=[]

for i in amounts:
    if i < 0:
        status= 'refunded'
        status_redone.append(status)
    else:
        status='completed'
        status_redone.append(status)
```

```python
data_sub['status']=status_redone
data_sub['status'].value_counts()
```
