```python
import numpy as np
import pandas as pd

np.random.seed(42)
```

```python
customer_data=pd.read_csv('customers.csv')
```

```python
customer_data.head()
```

```python
customer_data_sub=customer_data.copy()
```

```python
#there is inconsistency in the signup_date column
customer_data_sub['signup_date']=pd.to_datetime(customer_data_sub['signup_date'],format='mixed')
```

```python
customer_data_sub.info()
```

```python
missing_cols=customer_data_sub.isnull().any(axis=1).sum()
missing_rows=customer_data_sub.isnull().any(axis=0).sum()
missing_cols,missing_rows
```

```python
customer_data_sub['age'].describe()
```

```python
#filling nan in the age column with the distribution of age in the data
customer_data_sub['age']=customer_data['age'].fillna( np.random.randint(customer_data_sub['age'].min(), customer_data_sub['age'].max() ) )
```

```python
customer_data_sub['age'].describe()
```

```python
customer_data_sub['gender'].describe()
```

```python
customer_data_sub['gender'].value_counts()
```

```python
gender_encoding={'M':1, 'F':0}
```

```python
customer_data_sub['gender']=customer_data_sub['gender'].map(gender_encoding)
```

```python
customer_data_sub['gender'].value_counts()
```

```python
# fill nan of gender with either 1 or 0;  using random_distribution
customer_data_sub['gender']=customer_data_sub['gender'].fillna(pd.Series(np.random.randint(0,2,size=len(customer_data_sub))))
```

```python
customer_data_sub['gender'].value_counts()
```

```python
customer_data_sub['customer_id'].value_counts()
```

```python
customer_data_sub['email'].value_counts()
```

```python
customer_data_sub.head()
```

```python
# dropping nan for in the data;
customer_data_sub=customer_data_sub.dropna().reset_index()
customer_data_sub=customer_data_sub.drop(columns=['index'])
```

```python
customer_data_sub.info()
```

```python
customer_data_sub.head()
```

```python
customer_data_sub['customer_id'].value_counts()
```

```python
# changing the customer_id distribution since there is few unique entities along the dataset(C100,C_100,C101) which doesn't feel natural...
random_values=np.random.randint(1,(0.7 * len(customer_data_sub)),size=len(customer_data_sub))    
formatted_values=[f'C{i}'for i in random_values]
```

```python
# replacing existing values of customer_ids with new formatted_values
customer_data_sub['customer_id']=formatted_values
```

```python
customer_data_sub.head()
```

```python
# removing duplicates since they are there!
customer_data_sub=customer_data_sub.drop_duplicates()
```

```python
customer_data_sub.duplicated().sum()
```
